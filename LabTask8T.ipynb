```python
import math

def minimax_decision(current_depth, node_index, maximizing_player, leaf_values, depth_limit):
    if current_depth == depth_limit:
        return leaf_values[node_index]

    if maximizing_player:
        left_value = minimax_decision(current_depth + 1, node_index * 2, False, leaf_values, depth_limit)
        right_value = minimax_decision(current_depth + 1, node_index * 2 + 1, False, leaf_values, depth_limit)
        return max(left_value, right_value)
    else:
        left_value = minimax_decision(current_depth + 1, node_index * 2, True, leaf_values, depth_limit)
        right_value = minimax_decision(current_depth + 1, node_index * 2 + 1, True, leaf_values, depth_limit)
        return min(left_value, right_value)

leaf_nodes = [3, 5, 2, 9, 3, 5, 2, 9]
tree_depth = int(math.log(len(leaf_nodes), 2))

print("Optimal value:", minimax_decision(0, 0, True, leaf_nodes, tree_depth))
```
