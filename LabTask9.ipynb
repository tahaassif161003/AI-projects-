### Lab Task 9 
Write a function swapNodes(x, y) that takes two node values x and y, and swaps the nodes
containing these values.
- Do not swap data â€” swap the actual node links.
- No need to include a tail node
- Handle cases where:
- Where one or both values are not present.
- Where x and y are the same (no swap needed).
- Where one or both nodes is/are the head.
Ok let it go instead make a function swapNodes(x,y) (same name) but this function after changing the links of the previous nodes it will remove the nodes x and y from the linked list and end up not swapping then.
Also use functions insertAtEnd and display to insert and show the linked list.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insertAtEnd(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp

    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()

    def swapNodes(self, x, y):
        if x == y:
            return

        node_x = self.head
        node_y = self.head
        while node_x and node_x.data != x:
            node_x = node_x.next
        while node_y and node_y.data != y:
            node_y = node_y.next

        if not node_x or not node_y:
            return

        if node_x.prev:
            node_x.prev.next = node_x.next
        else:
            self.head = node_x.next
        if node_x.next:
            node_x.next.prev = node_x.prev

        if node_y.prev:
            node_y.prev.next = node_y.next
        else:
            self.head = node_y.next
        if node_y.next:
            node_y.next.prev = node_y.prev

dll = DoublyLinkedList()
for i in [1, 2, 3, 4, 5]:
    dll.insertAtEnd(i)

dll.display()
dll.swapNodes(2, 4)
dll.display()
```
