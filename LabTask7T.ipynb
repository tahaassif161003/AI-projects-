```python
network = {
    'A': {'B': 1, 'C': 3},
    'B': {'D': 1, 'E': 5},
    'C': {'F': 2},
    'D': {'G': 3},
    'E': {'G': 1},
    'F': {'G': 5},
    'G': {}
}

estimated_cost = {
    'A': 6,
    'B': 4,
    'C': 4,
    'D': 2,
    'E': 1,
    'F': 3,
    'G': 0
}

def a_star_search(start_node, end_node):
    frontier = {start_node: estimated_cost[start_node]}
    came_from = {start_node: None}
    actual_cost = {start_node: 0}

    while frontier:
        current_node = min(frontier, key=lambda node: actual_cost[node] + estimated_cost[node])
        print(f"Exploring Node: {current_node}")

        if current_node == end_node:
            route = []
            while current_node:
                route.append(current_node)
                current_node = came_from[current_node]
            return route[::-1]

        del frontier[current_node]

        for neighbor, travel_cost in network[current_node].items():
            tentative_g = actual_cost[current_node] + travel_cost
            if neighbor not in actual_cost or tentative_g < actual_cost[neighbor]:
                actual_cost[neighbor] = tentative_g
                frontier[neighbor] = tentative_g + estimated_cost[neighbor]
                came_from[neighbor] = current_node
    return None


destination = input("Enter the destination node: ")
result_path = a_star_search('A', destination)

if result_path:
    print("Path found:", result_path)
else:
    print("No path exists.")
```
