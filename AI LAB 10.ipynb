```python
df =df.replace(0, np.nan)
```

```python
null_values = np.sum(pd.isnull(df))
print("Total Null Values:", null_values)
```

```python
df['Pregnancies'] = df['Pregnancies'].fillna(99)
df['Glucose'] = df['Glucose'].fillna(101)
df['BloodPressure'] = df['BloodPressure'].fillna(65)
df['SkinThickness'] = df['SkinThickness'].fillna(34)
df["Insulin"] = df['Insulin'].fillna(100)
df['BMI'] = df['BMI'].fillna(23.8)
df['Outcome'] = df['Outcome'].fillna(1)
```

```python
df = df.drop('BMI',axis=1)
```

```python

x = df.iloc[:, 0:-1]
print("X shape:", x.shape)

y = df.iloc[:, -1]
print("Y shape:", y.shape)
```

```python
# Find all object columns
cat_columns = df.select_dtypes(['object']).columns

# Convert them into integers
df[cat_columns] = df[cat_columns].apply(lambda col: pd.factorize(col)[0])
```
